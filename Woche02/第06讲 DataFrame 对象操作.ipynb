### 6.1.5 `.loc[]` 索引访问方式

在DataFrame中，.loc[] 与 [] 有区别

- DataFrame 有两维，每一维都和 Series 的 `.loc[]` 用法相同；
- Series有四种方式，所以DataFrame有**16种**方式;
- 可以缺省后面维度，默认补全为 ':' 。

**下面都以第一维度为例，第二维可以类比。**

````python
heroes_df.loc['宋江','智慧'] # 返回指定元素值
````

````python
heroes_df.loc['宋江':'林冲','智慧'] # 返回序列切片
````

````python
heroes_df.loc[['宋江', '卢俊义', '林冲'],['忠诚', '智慧']] # 返回子数据帧
````

````python
heroes_df.loc[['宋江', '卢俊义', '林冲'],['忠诚']] # 返回子数据帧
````

### 6.1.6 `.iloc[]`，下标方式访问

- 注意：只用下标，不用索引
    - DataFrame 有两维，每一维都和 Series 的 `.iloc[]` 用法相同；
    - Series有四种方式，所以DataFrame有**16**种方式；
    - 可以缺省后面维度，默认补全为 `':'` 。

**下面都以第一维度为例，第二维可以类比。**

````python
heroes_df.iloc[1,1] # 返回指定元素值，卢俊义，智慧
````

````python
heroes_df.iloc[0:6,2]  # 返回子序列
````

````python
heroes_df.iloc[[0,1],[0,2]] # 返回子数据帧
````

````python
mask1 = [False, True, False, True, False, True, False, True]  # 返回子数据帧
mask2 = [True, False, True, False, False]
heroes_df.iloc[mask1,mask2]
````

## 6.2 修改

- 值 values
- 索引 index、columns

### 6.2.1 修改值 —— 访问右端赋值方式

- .loc[]方法确保原地修改

````python
# 修改前
df = heroes_df.copy()
df.loc['宋江']
````

````python
# 修改后
df.loc['宋江','勇敢'] -= 5 # 右端赋值
df.loc['宋江']
````

### 6.2.2 修改值 —— 函数批量方式


- 函数语法

```python
>>> pd.DataFrame.replace(to_replace=None, value=None, inplace=False)
```
- 参数说明
    - to_replace：拟修改的值，可以为列表
    - value：    修改后的值，可以为列表，与 to_repalce 要匹配
    - inplace：  是否在原地修改

````python
# 修改前
df = heroes_df.copy()
df
````

#### 修改特定值

````python
df.replace(to_replace = 99, value = 100, inplace=False)
````

````python
df
````

#### 交换两列

````python
df[['忠诚','智慧']] = df[['智慧','忠诚']]
df
````

### 6.2.3 修改索引 —— 直接方式

- 注意，索引类似于tuple
    - 必须完整修改
    - 不能部分修改

````python
# 修改前
df = heroes_df.copy()
df
````

````python
# 修改后
df.index = ['及时雨','玉麒麟', '智多星', '入云龙', '大刀', '豹子头', '霹雳火', '双鞭']
df.columns = ['忠','智','勇','技','命']
df
````

### 6.2.4 修改索引 —— 函数方式

- 语法

    ```python
    >>> pd.DataFrame.rename(index=None, columns=None,  level=None, inplace=False)
    ```

- 参数说明
    - index：list or dict，list时必须和已有索引长度相同，dict可以部分修改
    - columns：list or dict，list时必须长度相同，dict时可以部分修改
    - level：多重索引时，可以指定修改哪一重（暂不用）
    - inplace：是否原地修改。

````python
df = heroes_df.copy()
df.rename(index={'宋江':'黑宋江'}, columns={'忠诚':'忠义'}, inplace = False)
````

````python
df
````

## 6.3 增加行、列

### 6.3.1 增加单行 —— 字典方式

````python
df = heroes_df.copy()
df.loc['武松'] = [99,90,99,88,95]
df
````

### 6.3.2 增加多行 —— 函数方式

- 语法
    ```python
    >>> pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, 
                  keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)
    ```


- 参数说明
    - objs: 序列(Series)、数据帧(DataFrame)、面板(Panel)等对象为元素的列表、字典等系列(sequence)
    - axis： 0:沿行增加, 1：沿列增加，缺省为 0
    - 其它 (多数情况不须考虑，自学)
    

- 注意事项
    - 列索引 columns 要求一致
    - 行索引 index 避免重复

````python
df1 = heroes_df.copy()
df2 = pd.DataFrame(np.random.randint(90,100,(3,5)), index=["鲁智深","武松","李逵"], columns=["忠诚","智慧","勇敢","法术","生命"])

pd.concat([df1,df2], axis = 0)
````

### 6.3.3 增加单列 —— 字典方式

````python
df = heroes_df.copy()

df['秘技'] = np.random.randint(80,99,8)
df
````

### 6.3.4 增加多列 —— 函数方式

- 语法
    ```python
    >>> pd.concat(objs, axis=1, join='outer', join_axes=None, ignore_index=False, 
                  keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)
    ```


- 参数说明
    - objs: 序列(Series)、数据帧(DataFrame)、面板(Panel)等对象为元素的列表、字典等系列(sequence)
    - axis： 0:沿行增加, <font color='red'>1：沿列增加</font>，缺省为 0
    - 其它 (多数情况不须考虑，自学)
    

- 注意事项
    - 列索引 columns 要求一致
    - 行索引 index 避免重复

````python
df1 = heroes_df.copy()
df2 = pd.DataFrame(np.random.randint(90,100,(8,2)),
                   index=["宋江","卢俊义","吴用","公孙胜","关胜","林冲","秦明","呼延灼"],
                   columns=["密技","神力"])

pd.concat([df1,df2], axis=1)
````

## 6.4 删除

### 6.4.1 删除多行 —— 函数方式


- 语法
    ```python
    >>> pd.DataFrame.drop(self, labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
    ```


- 参数说明
    - labels: 单个标记、或标记列表，存放拟删除的行或列标记
    - axis： {0 or 'index', 1 or 'columns'}, default 0，是否删除行或列标记
    - inplace：是否本地修改
    - 其它 (多数情况不须考虑，自学)

````python
df = heroes_df.copy()
df
````

````python
df.drop(index=['宋江','吴用'], columns=['法术','生命'])
````

### 6.4.2 删除单列 —— 字典操作

````python
df = heroes_df.copy()
del df['法术']
df
````

### 6.4.3 删除多列 —— 函数方式


- 语法
    ```python
    >>> pd.DataFrame.drop(self, labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
    ```


- 参数说明
    - labels: 单个标记、或标记列表，存放拟删除的行或列标记
    - axis： 1 或 'columns', 删除列标记
